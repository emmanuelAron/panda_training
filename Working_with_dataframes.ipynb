```python
import pandas as pd
import numpy as np

#lists
names = ['Emma', 'Celine' , 'Bryan', 'Toto', 'Nicole', 'Tilda']
ages =  [44, 41, 25, 56, 33, 9]
married = [False,True , True , True , False , True]
```

```python
#Panda series
ser = pd.Series(names, name='name')
```

```python
ser
```

```python
#Panda dataframe
df = pd.DataFrame({'name' : names , 'age':ages , 'married':married})
```

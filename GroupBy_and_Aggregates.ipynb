```python


import pandas as pd
import numpy as np

games_url = 'https://andybek.com/pandas-games'
```

```python
games = pd.read_csv(games_url)
```

<font color='green'>Aggregates:</font>

```python
#Total sales across all regions?
games.loc[:,['NA_Sales' ,	'EU_Sales' ,	'JP_Sales' 	,'Other_Sales']]
```

```python
games.loc[:,['NA_Sales' ,	'EU_Sales' ,	'JP_Sales' 	,'Other_Sales']].sum()
```

```python
games.loc[:,['NA_Sales' ,	'EU_Sales' ,	'JP_Sales' 	,'Other_Sales']].max()
```

```python
#horizontal aggregation instead of vertical aggregation
games.loc[:,['NA_Sales' ,	'EU_Sales' ,	'JP_Sales' 	,'Other_Sales']].max(axis=1)
```

```python
#Find the total sales by region for X360 and PS3
sales = games.loc[:,['Platform','NA_Sales' ,	'EU_Sales' ,	'JP_Sales' 	,'Other_Sales']]
```
